```python
df['Memory'] = df['Memory'].astype(str).replace('\.0', '', regex=True)  # this is will remove .0
```

```python
df["Memory"] = df["Memory"].str.replace('GB', '')
df["Memory"] = df["Memory"].str.replace('TB', '000')
```

```python
df['Memory'].value_counts()
```

```python
new = df["Memory"].str.split("+", n = 1, expand = True)
'''
n=1: It specifies the maximum number of splits to perform. In this case,
it will perform the split only once at the first occurrence of '+'.
expand=True: It creates a DataFrame from the split elements.
If expand is True, the split elements are expanded into separate columns.
'''
```

```python
new
```

```python
df["first"]= new[0]
df["first"]=df["first"].str.strip()
```

```python
df["second"]= new[1]
df["second"]=df["second"].str.strip()
```

```python
df['first'].value_counts()
```

```python
df['Layer1SSD']=df['first'].apply(lambda x: 1 if 'SSD' in x else 0)
df['Layer1HDD']=df['first'].apply(lambda x: 1 if 'HDD' in x else 0)
df['Layer1FLASH_STORAGE']=df['first'].apply(lambda x: 1 if 'Flash Storage' in x else 0)
df['Layer1HYBRID']=df['first'].apply(lambda x: 1 if 'Hybrid' in x else 0)
```

```python
df['first'] = df['first'].str.replace(r'\D', '') ## it will remove all the non digit charaters
```

```python
df['first'].value_counts()
```

```python
df["second"].fillna("0", inplace = True)
```

```python
df["second"].value_counts()
```

```python
df['Layer2HDD']=df['second'].apply(lambda x: 1 if 'HDD' in x else 0)
df['Layer2SSD']=df['second'].apply(lambda x: 1 if 'SSD' in x else 0)
df['Layer2HYBRID']=df['second'].apply(lambda x: 1 if 'Hybrid' in x else 0)
```

```python
df['Layer2FLASH_STORAGE']=df['second'].apply(lambda x: 1 if 'Flash Storage' in x else 0)
```

```python
df['second'] = df['second'].str.replace(r'\D', '')  ## it will remove all the non digit charaters
```

```python
df['first']=df['first'].astype(int)
df['second']=df['second'].astype(int)
```

```python
df["HDD"]=(df["first"]*df["Layer1HDD"]+df["second"]*df["Layer2HDD"])
df["SSD"]=(df["first"]*df["Layer1SSD"]+df["second"]*df["Layer2SSD"])
df["Hybrid"]=(df["first"]*df["Layer1HYBRID"]+df["second"]*df["Layer2HYBRID"])
df["Flash_Storage"]=(df["first"]*df["Layer1FLASH_STORAGE"]+df["second"]*df["Layer2FLASH_STORAGE"])
```

```python
df.drop(columns=['first', 'second', 'Layer1HDD', 'Layer1SSD', 'Layer1HYBRID',
       'Layer1FLASH_STORAGE', 'Layer2HDD', 'Layer2SSD', 'Layer2HYBRID',
       'Layer2FLASH_STORAGE'],inplace=True)
```

```python
cat_columns=[i for i in X.columns if X[i].dtype=='object']
num_columns=[i for i in X.columns if X[i].dtype!='object']
```

```python
cat_columns
```

```python
num_columns
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
numeric_transformer=StandardScaler()
ohe=OneHotEncoder()
```

```python
preprocessor=ColumnTransformer([('numeric_transformer',numeric_transformer,num_columns),('ohe',ohe,num_columns)])
```

```python
X_pre=preprocessor.fit_transform(X)
```

```python
X_pre
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X_pre,y,test_size=0.2,random_state=42)
```

```python
X_train.shape,X_test.shape
```

```python
from sklearn.linear_model import LinearRegression, Lasso, Ridge
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.neighbors import KNeighborsRegressor
from xgboost import XGBRegressor
from sklearn.model_selection import RandomizedSearchCV
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
```

```python
models={'LinearRegression ':LinearRegression(),'Lasso':Lasso(),'Ridge':Ridge(),'Random_Forest_Regressor':RandomForestRegressor(),
        'Gradient_Boosting_Regression':GradientBoostingRegressor(),'Decision_Tree_Regressor':DecisionTreeRegressor(),
        'XGBOOST':XGBRegressor()
}
```

```python
for i in range(len(models)):
    model=list(models.values())[i]
    model.fit(X_train,y_train)
    y_pred=model.predict(X_test)
    R2_SCORE=r2_score(y_test,y_pred)
    print(f'{list(models.keys())[i]} with R2 SCORE: {R2_SCORE}')
```
